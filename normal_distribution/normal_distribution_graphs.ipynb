# Normal Distribution Graphs

### Playing with plots of the Normal Distribution

###### The idea here is to see the effect of changing the standard deviation and the mean on the graphs
###### DISCLAIMER: This may not render correctly if you are viewing on Github. For such cases, this notebook must be run locally

```python
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, truncnorm
#import ipywidgets as widgets # integration of interactive widgets
#from ipywidgets import interact, interact_manual
```

###### Creating dummy array for the 1st plotted curve

```python
mean1 = 126
sd1 = 6
x = np.random.normal(loc=mean1, scale=sd1, size=1000)
#x = truncnorm.rvs(a=0, b=1, loc=mean1, scale=sd1, size=1000)
x = np.sort(x)
```

###### Creating dummy array for the 2nd plotted curve

```python
mean2 = 126
sd2 = 1.5
y = np.random.normal(loc=mean2, scale=sd2, size=1000)
#y = truncnorm.rvs(a=0, b=1, loc=mean2, scale=sd2, size=1000)
y = np.sort(y)
```

###### Creating dummy array for the 3rd plotted curve

```python
mean3 = 85
sd3 = 6
z = np.random.normal(loc=mean3, scale=sd3, size=1000)
#z = truncnorm.rvs(a=0, b=1, loc=mean3, scale=sd3, size=1000)
z = np.sort(z)
```

```python
#Apply pdf function to the data
#pdf = normal_dist(x,mean,sd)
pdf1 = norm.pdf(x, loc=mean1, scale=sd1)
pdf2 = norm.pdf(y, loc=mean2, scale=sd2)
pdf3 = norm.pdf(z, loc=mean3, scale=sd3)
 
#Plotting the Results
plt.figure(figsize=(10, 6))
line1 = plt.plot(x, pdf1, color = 'red', label='Random sample, μ=126, σ=6')#, marker='x', linestyle='')
line2 = plt.plot(y, pdf2, color = 'blue', label='Random sample, μ=126, σ=1.5')
line3 = plt.plot(z, pdf3, color = 'green', label='Random sample, μ=85, σ=6')
plt.xlabel('Random data points')
plt.ylabel('Probability Density')
plt.legend()
```

```python
#Using CDF function on the data
#pdf = normal_dist(x,mean,sd)
cdf1 = norm.cdf(x, loc=mean1, scale=sd1)
cdf2 = norm.cdf(y, loc=mean2, scale=sd2)
cdf3 = norm.cdf(z, loc=mean3, scale=sd3)

#Plotting the CDF graph
plt.figure(figsize=(10, 6))
plt.plot(x, cdf1, color = 'red', label='Random sample, μ=126, σ=6')
plt.plot(y, cdf2, color = 'blue', label='Random sample, μ=126, σ=1.5')
plt.plot(z, cdf3, color = 'green', label='Random sample, μ=85, σ=6')
plt.xlabel('Random data points')
plt.ylabel('Probability')
plt.legend()
```
