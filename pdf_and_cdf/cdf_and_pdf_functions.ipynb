# PDF and CDF

### Using the pdf and cdf functions available in Python

###### Key to the pdf and cdf functions working, the data series must be ordered either ascending or descending but cannot be in a randomised format otherwise the plots (both cdf and pdf) will be coverd in chaotic lines rather than one smooth plot

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#data = pd.read_csv('./SOCR-HeightWeight.csv')
data = pd.read_csv('./players_stats.csv')
```

```python
data
```

```python
x = data[['Height']].sort_values(by='Height', ascending=True)
#x = data[['Height(Inches)']].sort_values(by='Height(Inches)', ascending=True)
#x = data[['Weight(Pounds)']].sort_values(by='Weight(Pounds)', ascending=True)
#x = np.linspace(1,50,200)
```

```python
x
```

```python
#Function which manually calculates the pdf of the normal distribution
def normal_dist(x , mean , sd):
    prob_density = (np.pi*sd) * np.exp(-0.5*((x-mean)/sd)**2)
    return prob_density
```

```python
#Calculate mean and Standard deviation
mean = np.mean(x)
sd = np.std(x)
```

```python
#Apply pdf function to the data
#pdf = normal_dist(x,mean,sd)
pdf = norm.pdf(x, loc=mean, scale=sd)
 
#Plotting the Results
plt.plot(x,pdf , color = 'red')#, marker='x', linestyle='')
plt.xlabel('Data points')
plt.ylabel('Probability Density')
```

```python
#Using CDF function on the data
#pdf = normal_dist(x,mean,sd)
cdf = norm.cdf(x, loc=mean, scale=sd)

#Plotting the CDF graph
plt.plot(x,cdf , color = 'red')
plt.xlabel('Data points')
plt.ylabel('Probability')
```

```python
#Using CDF to find probability of specific cases
cdf =  norm(loc=mean, scale=sd).cdf(20)
```
